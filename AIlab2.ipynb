Implement a Simple Reflex Agent for a vacuum cleaner that can navigate a three-room environment (A, B, and C) and maintain cleanliness. Identify the “rationality” of its actions. How will you define the performance cost? Define the complete Rule Table and store it either in a dictionary/vector/matrix based on the language you implement the vacuum cleaner. Show the simulation output with percept, action, and location. Do you need priorities in the rules?

```python
# -----------------------------------------
# Simple Reflex Agent: Vacuum Cleaner
# Environment: Three rooms (A, B, C)
# Agent has NO memory (pure reflex)
# -----------------------------------------

# Initial environment state
environment = {
    "A": "Dirty",
    "B": "Dirty",
    "C": "Clean"
}

# Agent starting location
agent_location = "A"

# Rule table (Percept -> Action)
rules = {
    ("A", "Dirty"): "Clean",
    ("A", "Clean"): "Move_Right",
    ("B", "Dirty"): "Clean",
    ("B", "Clean"): "Move_Right",
    ("C", "Dirty"): "Clean",
    ("C", "Clean"): "Move_Left"
}

# Movement logic
def move(location, action):
    if action == "Move_Right":
        if location == "A":
            return "B"
        elif location == "B":
            return "C"
    elif action == "Move_Left":
        if location == "C":
            return "B"
        elif location == "B":
            return "A"
    return location

print("VACUUM CLEANER AGENT SIMULATION")
print("-" * 50)
print("Step | Location | Percept | Action")
print("-" * 50)

step = 1

# Run simulation
for _ in range(6):
    percept = (agent_location, environment[agent_location])
    action = rules[percept]

    print(f"{step:>4} | {agent_location:^8} | {environment[agent_location]:^7} | {action}")

    if action == "Clean":
        environment[agent_location] = "Clean"
    else:
        agent_location = move(agent_location, action)

    step += 1


print("-" * 50)
print("All rooms are clean. Agent stops (NoOp).")
```

2. Design a Simple Reflex Agent for an Indian Railways Level Crossing. The agent must
ensure maximum safety for road traffic while minimizing delays for the train, using
multi-source sensory input without historical tracking.
The environment consists of a double-track railway line intersecting a busy state
highway. The percepts include the following:
● Track Sensors (Inbound/Outbound): Sensors placed 2km away that detect if a
train is currently passing over them. It is either detected or not detected.
● Obstacle Sensors (Yellow Box): A grid of sensors on the road between the gates
to detect “stuck” vehicles (e.g., an auto-rickshaw or buffalo).
● Manual Emergency Input: A physical lever for the station master. It is either
Neutral or Active.
Actuators:
● Gate Arm: Lower/Raise
● Hooter/Siren: On/Off
● Signal to Train: Green (Safe to proceed)/Red (Emergency Stop)

Design the rules set. Show the simulation output with percept, action, and location. Do you need
priorities in the rules?

```python
# --------------------------------------------------
# Simple Reflex Agent: Indian Railways Level Crossing
# No memory, no historical tracking
# Decision based ONLY on current percepts
# --------------------------------------------------

# Agent function
def level_crossing_agent(inbound_train, outbound_train, obstacle, emergency):
    """
    inbound_train  : True / False
    outbound_train : True / False
    obstacle       : True / False
    emergency      : True / False
    """

    # PRIORITY-BASED RULES (Safety First)

    if emergency:
        return "Lower Gate", "Hooter ON", "Signal RED"

    if obstacle:
        return "Lower Gate", "Hooter ON", "Signal RED"

    if inbound_train or outbound_train:
        return "Lower Gate", "Hooter ON", "Signal GREEN"

    return "Raise Gate", "Hooter OFF", "Signal GREEN"


# --------------------------------------------------
# Simulation Scenarios
# --------------------------------------------------

scenarios = [
    # Scenario Name, Inbound, Outbound, Obstacle, Emergency
    ("Train Approaching (Inbound)", True, False, False, False),
    ("Train Approaching (Outbound)", False, True, False, False),
    ("Vehicle Stuck on Crossing", False, False, True, False),
    ("Emergency Lever Activated", False, False, False, True),
    ("All Clear", False, False, False, False)
]

print("INDIAN RAILWAYS LEVEL CROSSING SIMULATION")
print("-" * 95)
print("Scenario | Inbound | Outbound | Obstacle | Emergency | Gate | Hooter | Signal")
print("-" * 95)

# Run simulation
for s in scenarios:
    name, inbound, outbound, obstacle, emergency = s
    gate, hooter, signal = level_crossing_agent(inbound, outbound, obstacle, emergency)

    print(f"{name:<28} | {str(inbound):<7} | {str(outbound):<8} | {str(obstacle):<8} | "
          f"{str(emergency):<9} | {gate:<10} | {hooter:<9} | {signal}")

print("-" * 95)
```
